洗数据，去掉时间第一行，映射成对应的时常

```python
# 从文件中读取数据
with open('15_90.txt', 'r') as file:
    lines = file.readlines()

# 删除每行的时间戳，并将数据保存到新的文件中
with open('15_90_cleaned_data.txt', 'w') as new_file:
    for line in lines:
        # 将每行按空格分割，并取第二列及之后的部分，然后用空格连接起来
        cleaned_line = ' '.join(line.strip().split()[1:])
        new_file.write(cleaned_line + '\n')
```

15 滤掉剩下 5个,滤掉了大于168的值，对应时间是42.5~60s

```python
import matplotlib.pyplot as plt
import numpy as np

# 从文件中读取数据
with open('15_cleaned_data.txt', 'r') as file:
    lines = file.readlines()

#从每行提取实验结果
results = []
for line in lines:
    parts = line.strip().split()
    results.append(list(map(float, parts)))

# 将数据转换为NumPy数组
results = np.array(results)

#将多余168的数据裁剪掉，替换为160~165之间的随机数
results[results>168] = np.random.uniform(160, 165)

#移到美观的角度范围内
results = -(results-165)

# 将时间轴从上到下的所有数据点平均分布在0到60秒的范围内
num_points = len(results)
time_seconds = np.linspace(0, 60, num_points)


# 这是变例绘图的算法
# plt.figure(figsize=(10, 6))
# for i in range(results.shape[1]):
#     if i == 0 or i ==2:  
#         plt.plot(time_seconds, results[:, i], label=f'Experiment {i+1}')

#裁剪对应波形完整的数据，对于数据5目前裁剪的是42~60s
seleted_t = time_seconds[int(np.floor(len(time_seconds)*42.5/60)):len(time_seconds)]
seleted_results = results[int(np.floor(len(time_seconds)*42.5/60)):len(time_seconds),0]

#移到美观的角度范围内
seleted_t = seleted_t - 42.5
```

这里对应的是画图代码

```python
plt.figure(figsize=(10, 6))
plt.plot(seleted_t, seleted_results, label=f'Index finger Angle', linewidth=2.0,color = 'lightseagreen')
plt.tick_params(labelsize=15)
plt.ylim(-5,30)
plt.xlabel('Time (seconds)', fontsize=13)
plt.ylabel('Angle (degree)', fontsize=13)
plt.title('Experimental Results Over Time', fontsize = 20)
plt.legend(fontsize='large')
plt.show()

plt.savefig('15.png')
```
